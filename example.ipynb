```python
from utils import load_ssat_model, load_face_parser
from ssat import transfer

from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

### Loading models

```python
ssat_path = "models/SSAT.pth"
face_parsing_path = "models/PARSING.pth"
device = "cpu"

ssat = load_ssat_model(ssat_path, device)
parser = load_face_parser(face_parsing_path, device)
```

### Loading images

```python
source = Image.open("images/non-makeup.png")
target = Image.open("images/makeup.png")

fig, ax = plt.subplots(1, 2, figsize=(16, 9))
ax[0].axis("off")
ax[0].imshow(np.asarray(source))
ax[1].axis("off")
ax[1].imshow(np.asarray(target))
```

### Transfer makeup style

```python
result = transfer(ssat, parser, source, target, speed=False).numpy()
result = (result * 255).astype(np.uint8).transpose(1, 2, 0)
image = Image.fromarray(result, "RGB")
```
